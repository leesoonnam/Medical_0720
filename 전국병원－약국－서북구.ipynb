# 수거약품목 추가하는 방법
1) CSV 파일 열기
2) 수거약품목으로 쓸 ‘약품목 리스트’ 만들기
3) 각 약국에 랜덤으로 1~3개 선택해서 추가하기
4) 새 CSV로 저장하기


```python
# ----------------------------------------
# 📌 1) pandas와 random 도구 꺼내오기
# ----------------------------------------
import pandas as pd      # pandas는 표(데이터프레임)를 다루는 도구!
import random            # random은 랜덤으로 뽑을 때 사용하는 도구!

# ----------------------------------------
# 📌 2) CSV 파일 불러오기
# ----------------------------------------
# 'cheonan_seobuk_pharmacy.csv' 파일이 같은 폴더에 있어야 해요!
# encoding='utf-8-sig'로 열면 한글 깨짐 방지!
file_path = "cheonan_seobuk_pharmacy.csv"
df = pd.read_csv(file_path, encoding="utf-8-sig")

print("\n✅ 원본 데이터 미리보기")
print(df.head())  # 데이터 앞부분 5행만 살짝 보기

# ----------------------------------------
# 📌 3) 예시 약품목 카테고리 리스트 만들기
# ----------------------------------------
# 각 약국에 랜덤으로 붙여줄 '수거약품목' 메뉴판!
categories = [
    "해열제", "진통제", "소염제", "항생제", "소화제",
    "심혈관계", "호흡기계", "내분비계", "신경계",
    "안이비인후과용", "외용제", "건강기능식품"
]

# ----------------------------------------
# 📌 4) 랜덤으로 1~3개 약품목 뽑아주는 함수 만들기
# ----------------------------------------
# random.sample(목록, 개수) → 목록에서 중복 없이 무작위로 뽑음!
# ", ".join()으로 쉼표로 연결해 문자열로 만들어줌!
def random_items():
    return ", ".join(random.sample(categories, random.randint(1, 3)))

# ----------------------------------------
# 📌 5) 각 약국마다 '수거약품목' 컬럼 추가하기
# ----------------------------------------
# df.apply(함수, axis=1)는 약국 한 줄씩 돌면서 함수를 실행!
df['수거약품목'] = df.apply(lambda row: random_items(), axis=1)

print("\n✅ 수거약품목 컬럼 추가 완료!")
print(df[['병원명', '수거약품목']].head())  # 병원명 + 수거약품목만 보기

# ----------------------------------------
# 📌 6) 새로운 CSV 파일로 저장하기
# ----------------------------------------
# index=False : 행 번호는 저장하지 않음
# encoding='utf-8-sig' : 한글 깨짐 방지!
output_path = "cheonan_seobuk_pharmacy_with_items.csv"
df.to_csv(output_path, index=False, encoding="utf-8-sig")

print(f"\n✅ 새 파일 저장 완료: {output_path}")
print("이제 이 파일로 체크박스 필터링 & 지도 실습에 쓰면 됩니다!")
```
